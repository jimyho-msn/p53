```python
def cut_word(word,method):
    position = word.find ("-")
    if position != -1:
        bottomsalary = word[:position-1]
        topsalary = word[position+1:len(word)-1]
    else:
        bottomsalary = word[:word.upper().find("K")]
        topsalary = bottomsalary
    if method == "bottom":
        return bottomsalary
    else:
        return topsalary
```

```python
# 规范化，最低薪酬，变成数字
df_duplicates["bottomsalary"] = df_duplicates.salary.apply(cut_word,method = "bottom") 
```

```python
df_duplicates["bottomsalary"].head()
```

```python
df_duplicates["topsalary"] = df_duplicates.salary.apply(cut_word,method = "top") 
```

```python
df_duplicates.topsalary.head(2)
```

```python
df_duplicates.bottomsalary.astype("int") # 仅是显示，并未赋值给 df_duplicates
```

```python
df_duplicates.topsalary.astype("int") # 仅是显示，并未赋值给 df_duplicates
```

```python
# 赋值给 df_duplicates
df_duplicates["bottomsalary"] = df_duplicates.bottomsalary.astype("int")
df_duplicates["topsalary"] = df_duplicates.topsalary.astype("int")
```

```python
df_duplicates["avgsalary"]=df_duplicates.apply(lambda x:(x.bottomsalary+x.topsalary)/2,axis=1)
```

```python
# agg 可以分组，组内排序，但不能取前N名，只能全部列出
#分组后，组内排序，取前N名 
# x : 数据列, n : 前几
def topN(x,n):
    counts = x.value_counts()
    r = counts.sort_values(ascending=False)
    return r[:n]
# 按城市分组，组内公司招聘个数排序，前5名
df_clean.groupby("city").companyShortName.apply(topN,n = 5).head(20)
```

```python
# 按城市分组，组内职位招聘个数排序，前5名
df_clean.groupby("city").positionName.apply(topN,n = 5).head(20)
```

```python
# 按城市分组，组内按工作年限招聘个数排序，前5名
df_clean.groupby("city").workYear.apply(topN,n = 5).head(20)
```

```python
def topSalaryN(x,n):
    r = x.sort_values("avgsalary",ascending=False)
    return r[:n]
```

```python
#按城市分组，薪酬排名前5，avgsalary 由高到低
#没有指定返回列，所有返回所有列
df_clean.groupby("city").apply(topSalaryN,n=5)
```

```python
#指定返回列
df_clean.groupby("city").apply(topSalaryN,n=5)[["companyShortName","city","avgsalary"]] 
```

```python
#深加工，认为分出level,  新增 level 列，将数值型数据加工成分类型数据。加工成标签。
#cut 用作分类
bins = [0,3,5,10,15,20,30,100]
level = ["0-3","3-5","5-10","10-15","15-20","20-30","30+"]
df_clean["level"] = pd.cut(df_clean.avgsalary,bins,labels = level)
```

```python
df_clean[["avgsalary","level"]]
```

```python
#堆积百分比柱形图，用上了level 列
df_level = df_clean.groupby(["city","level"]).avgsalary.count().unstack()
```

```python
df_level.head()
```

```python
#用 lambda 转换成占总数百分比
df_level_prop = df_level.apply(lambda x: x/x.sum(),axis = 1)
```

```python
df_level_prop.head()
```
